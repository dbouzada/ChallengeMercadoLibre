```python
import yfinance as yf
import pandas as pd
from datetime import datetime

# Diccionario de IDs, tickers, nombres, países, continentes y sectores de las fintechs
fintechs = {
    1: ("PYPL", "PayPal", "USA", "North America", "Digital Payments"),
    2: ("SQ", "Square", "USA", "North America", "Digital Payments"),
    3: ("V", "Visa", "USA", "North America", "Digital Payments"),
    4: ("MA", "Mastercard", "USA", "North America", "Digital Payments"),
    5: ("INTU", "Intuit", "USA", "North America", "Financial Software"),
    6: ("ADYEN", "Adyen", "Netherlands", "Europe", "Digital Payments"),
    7: ("MELI", "Mercado Libre", "Argentina", "South America", "E-commerce/Fintech"),
    8: ("NU", "Nubank", "Brazil", "South America", "Digital Banking"),
    9: ("SHOP", "Shopify", "Canada", "North America", "E-commerce"),
    10: ("FIS", "Fidelity National Information Services", "USA", "North America", "Financial Services"),
    11: ("FISV", "Fiserv", "USA", "North America", "Financial Services"),
    12: ("GPN", "Global Payments", "USA", "North America", "Digital Payments"),
    13: ("BR", "Broadridge Financial Solutions", "USA", "North America", "Financial Services"),
    14: ("SSNC", "SS&C Technologies", "USA", "North America", "Financial Services"),
    15: ("FICO", "Fair Isaac", "USA", "North America", "Financial Software"),
    16: ("JKHY", "Jack Henry & Associates", "USA", "North America", "Financial Services"),
    17: ("FLT", "FleetCor", "USA", "North America", "Digital Payments"),
    18: ("EVTC", "Evertec", "Puerto Rico", "North America", "Digital Payments"),
    19: ("JPM", "JPMorgan Chase", "USA", "North America", "Banking"),
    20: ("C", "Citigroup", "USA", "North America", "Banking"),
    21: ("AXP", "American Express", "USA", "North America", "Digital Payments"),
    22: ("SYF", "Synchrony Financial", "USA", "North America", "Financial Services"),
    23: ("COF", "Capital One", "USA", "North America", "Banking"),
    24: ("DFS", "Discover Financial", "USA", "North America", "Financial Services"),
    25: ("ALLY", "Ally Financial", "USA", "North America", "Digital Banking"),
    26: ("TROW", "T. Rowe Price", "USA", "North America", "Asset Management"),
    27: ("BLK", "BlackRock", "USA", "North America", "Asset Management"),
    28: ("MS", "Morgan Stanley", "USA", "North America", "Banking/Financial Services"),
    29: ("GS", "Goldman Sachs", "USA", "North America", "Banking/Financial Services"),
    30: ("AMTD", "TD Ameritrade", "USA", "North America", "Brokerage"),
    31: ("SCHW", "Charles Schwab", "USA", "North America", "Brokerage"),
    32: ("IBKR", "Interactive Brokers", "USA", "North America", "Brokerage"),
    33: ("ETFC", "E*TRADE", "USA", "North America", "Brokerage"),
    34: ("MTCH", "Match Group", "USA", "North America", "Online Services"),
    35: ("DLO", "DLocal", "Uruguay", "South America", "Digital Payments"),
    36: ("STNE", "StoneCo", "Brazil", "South America", "Digital Payments"),
    37: ("AFRM", "Affirm", "USA", "North America", "Buy Now, Pay Later"),
    38: ("SOFI", "SoFi", "USA", "North America", "Personal Finance"),
    39: ("HOOD", "Robinhood", "USA", "North America", "Brokerage"),
    40: ("BILL", "Bill.com", "USA", "North America", "Financial Software"),
    41: ("LC", "LendingClub", "USA", "North America", "Peer-to-Peer Lending"),
    42: ("UPST", "Upstart Holdings", "USA", "North America", "Lending"),
    43: ("XLF", "Financial Select Sector SPDR", "USA", "North America", "Financial Services ETF"),
    44: ("ARCC", "Ares Capital", "USA", "North America", "Private Equity"),
    45: ("BABA", "Alibaba", "China", "Asia", "E-commerce/Fintech"),
    46: ("NTNX", "Nutanix", "USA", "North America", "Cloud Computing"),
    47: ("TWLO", "Twilio", "USA", "North America", "Cloud Communications"),
    48: ("ZBRA", "Zebra Technologies", "USA", "North America", "Technology"),
    49: ("WU", "Western Union", "USA", "North America", "Digital Payments"),
    50: ("MC", "Moelis & Company", "USA", "North America", "Investment Banking"),
    51: ("TD", "Toronto-Dominion Bank", "Canada", "North America", "Banking"),
    52: ("BMO", "Bank of Montreal", "Canada", "North America", "Banking"),
    53: ("RY", "Royal Bank of Canada", "Canada", "North America", "Banking"),
    54: ("CM", "Canadian Imperial Bank of Commerce", "Canada", "North America", "Banking"),
    55: ("BNS", "Bank of Nova Scotia", "Canada", "North America", "Banking"),
    56: ("CS", "Credit Suisse", "Switzerland", "Europe", "Banking"),
    57: ("UBS", "UBS Group", "Switzerland", "Europe", "Banking"),
    58: ("SAN", "Banco Santander", "Spain", "Europe", "Banking"),
    59: ("HSBC", "HSBC Holdings", "UK", "Europe", "Banking"),
    60: ("BBVA", "BBVA", "Spain", "Europe", "Banking"),
    61: ("ITUB", "Itaú Unibanco", "Brazil", "South America", "Banking"),
    62: ("PAGS", "PagSeguro", "Brazil", "South America", "Digital Payments"),
    63: ("PDD", "Pinduoduo", "China", "Asia", "E-commerce"),
    64: ("NTES", "NetEase", "China", "Asia", "Technology"),
    65: ("BILI", "Bilibili", "China", "Asia", "Entertainment"),
    66: ("JD", "JD.com", "China", "Asia", "E-commerce"),
    67: ("PBR", "Petrobras", "Brazil", "South America", "Energy"),
    68: ("VALE", "Vale", "Brazil", "South America", "Mining"),
    69: ("TM", "Toyota Motor", "Japan", "Asia", "Automotive"),
    70: ("RACE", "Ferrari", "Italy", "Europe", "Automotive"),
    71: ("UGP", "Ultrapar", "Brazil", "South America", "Energy"),
    72: ("YPF", "YPF", "Argentina", "South America", "Energy"),
    73: ("MSFT", "Microsoft", "USA", "North America", "Technology"),
    74: ("AAPL", "Apple", "USA", "North America", "Technology"),
    75: ("GOOGL", "Alphabet", "USA", "North America", "Technology"),
    76: ("AMZN", "Amazon", "USA", "North America", "E-commerce"),
    77: ("FB", "Meta Platforms", "USA", "North America", "Technology"),
    78: ("TSLA", "Tesla", "USA", "North America", "Automotive"),
    79: ("NFLX", "Netflix", "USA", "North America", "Entertainment"),
    80: ("NVDA", "NVIDIA", "USA", "North America", "Technology"),
    81: ("ADBE", "Adobe", "USA", "North America", "Technology"),
    82: ("CRM", "Salesforce", "USA", "North America", "Technology"),
    83: ("INTC", "Intel", "USA", "North America", "Technology"),
    84: ("CSCO", "Cisco", "USA", "North America", "Technology"),
    85: ("AMD", "Advanced Micro Devices", "USA", "North America", "Technology"),
    86: ("SQM", "Sociedad Química y Minera", "Chile", "South America", "Chemicals"),
    87: ("LFC", "China Life Insurance", "China", "Asia", "Insurance"),
    88: ("JD", "JD.com", "China", "Asia", "E-commerce"),
    89: ("PDD", "Pinduoduo", "China", "Asia", "E-commerce"),
    90: ("BABA", "Alibaba", "China", "Asia", "E-commerce"),
    91: ("WUBA", "58.com", "China", "Asia", "Technology"),
    92: ("ATHM", "Autohome", "China", "Asia", "Technology"),
    93: ("BZUN", "Baozun", "China", "Asia", "E-commerce"),
    94: ("TME", "Tencent Music Entertainment", "China", "Asia", "Entertainment"),
    95: ("YY", "JOYY", "China", "Asia", "Entertainment"),
    96: ("QFIN", "360 DigiTech", "China", "Asia", "Fintech"),
    97: ("FUTU", "Futu Holdings", "China", "Asia", "Brokerage"),
    98: ("AMTD", "AMTD International", "Hong Kong", "Asia", "Financial Services"),
    99: ("NIO", "NIO Inc.", "China", "Asia", "Automotive"),
    100: ("XPEV", "XPeng Inc.", "China", "Asia", "Automotive")
}

# Establecer el rango de fechas desde 2019 hasta hoy
start_date = "2019-01-01"
end_date = datetime.today().strftime('%Y-%m-%d')

# Descargar los datos
data = yf.download(tickers=[fintech[0] for fintech in fintechs.values()], start=start_date, end=end_date)

# Calcular la capitalización de mercado (Market Cap) y otros datos
market_cap_data = []
for id_, (ticker, name, country, continent, sector) in fintechs.items():
    ticker_data = yf.Ticker(ticker)
    company_data = data['Adj Close'][ticker].dropna()  # Remover datos faltantes

    # Obtener el número de acciones en circulación y otros indicadores
    shares_outstanding = ticker_data.info.get('sharesOutstanding')
    pe_ratio = ticker_data.info.get('forwardPE')
    dividend_yield = ticker_data.info.get('dividendYield')
    beta = ticker_data.info.get('beta')

    if shares_outstanding:
        for date, price in company_data.items():
            market_cap = price * shares_outstanding
            market_cap_data.append({
                'ID': id_,
                'Ticker': ticker,
                'Company_Name': name,
                'Country': country,
                'Continent': continent,
                'Date': date.strftime('%Y-%m-%d'),  # Formato YYYY-MM-DD
                'Market_Cap': market_cap,
                'Close_Price': price,
                'PE_Ratio': pe_ratio,
                'Dividend_Yield': dividend_yield,
                'Beta': beta,
                'Sector': sector,  
            })

# Crear el DataFrame
market_cap_df = pd.DataFrame(market_cap_data)
market_cap_df['Date'] = pd.to_datetime(market_cap_df['Date'], format='%Y-%m-%d')

# Mostrar el DataFrame como tabla en Python
print(market_cap_df)

# Guardar el DataFrame como archivo CSV
csv_file_path = 'fintechs_data.csv'
market_cap_df.to_csv(r'C:\Users\dbouz\Downloads\fintechs_data.csv', index=False)

# Mostrar el DataFrame como tabla en Python
market_cap_df

print(f"Archivo CSV guardado en: {csv_file_path}")
```
